```python
#ho="There is significant difference in the average price of Diesel and Petrol cars."
#ha="There is no significant difference in the average price of Diesel and Petrol cars."
```

```python

data['fuelType'].value_counts()
```

```python
one=data[data['fuelType']=='Diesel']['price']
two=data[data['fuelType']=='Petrol']['price']
                         
```

```python
from scipy.stats import ttest_ind
_,pvalue=ttest_ind(one,two,equal_var=False)
print(pvalue)
```

```python
if pvalue < 0.05:
    print('The average price of Diesel and Petrol cars is significantly different.')
else:
    print('"There is no significant difference in the average price of Diesel and Petrol cars.')
```

# anova test

```python
#To perform an ANOVA test on the Audi dataset to compare the average price of used cars based on different 
#fuel types
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

formula = 'price ~ C(fuelType)'
model = ols(formula, data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
```
